```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_paths(csv_file='/Users/zhoupeng/Desktop/CF_coursework/data/paths_data.csv'):
    df = pd.read_csv(csv_file)
    
    plt.figure(figsize=(12, 8))
    
    # Plot price paths
    path_columns = [col for col in df.columns if 'Path' in col]
    for col in path_columns:
        plt.plot(df['Step'], df[col], alpha=0.6, label=col)
    
    # Plot barrier and strike
    plt.axhline(y=df['Strike'].iloc[0], color='r', linestyle='--', label='Strike')
    plt.plot(df['Step'], df['Barrier'], 'k--', label='Barrier')
    
    plt.xlabel('Time Steps')
    plt.ylabel('Stock Price')
    plt.title('Monte Carlo Price Paths')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_paths()
```
